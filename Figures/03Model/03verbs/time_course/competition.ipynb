```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
```

```python
transport = np.genfromtxt('recording_2022_01_18__12_13_24/TRANSPORT[OUTPUT].sigmoided_activation.csv', skip_header=1, delimiter=',')[:,1]
drop = np.genfromtxt('recording_2022_01_18__12_13_24/DROP[OUTPUT].sigmoided_activation.csv', skip_header=1, delimiter=',')[:,1]
grasp = np.genfromtxt('recording_2022_01_18__12_13_24/GRASP[OUTPUT].sigmoided_activation.csv', skip_header=1, delimiter=',')[:,1]
move = np.genfromtxt('recording_2022_01_18__12_13_24/transport[OUTPUT].boost.csv', skip_header=1, delimiter=',')[:,1]
pull = np.genfromtxt('recording_2022_01_18__12_13_24/grasp[OUTPUT].boost.csv', skip_header=1, delimiter=',')[:,1]
push = np.genfromtxt('recording_2022_01_18__12_13_24/drop[OUTPUT].boost.csv', skip_header=1, delimiter=',')[:,1]
size_change = np.genfromtxt('recording_2022_01_18__12_13_24/size_change[OUTPUT].boost.csv', skip_header=1, delimiter=',')[:,1]
```

```python
transport.shape
```

```python
plt.rcParams.update({'font.size': 40})
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(20,8))

ax[0].plot(transport, label='$transport$', linestyle='-', color='black')
ax[0].plot(grasp, label='$grasp$', linestyle='-', color='green')
ax[0].plot(drop, label='$drop$', linestyle='-', color='blue')
ax[0].legend(bbox_to_anchor=(0.95, 1.1), frameon=False)
ax[0].set(xticks=[], ylabel='activation')

ax[1].plot(move, label='motion direction', linestyle='--', color='black')
ax[1].plot(push, label='push', linestyle='--', color='blue')
ax[1].plot(size_change, label='size change', linestyle='--', color='cyan')
ax[1].plot(pull, label='pull', linestyle='--', color='green')
ax[1].set(xticks=[], yticklabels=['',0 , ''], xlabel='time', ylabel='activation')
ax[1].legend(bbox_to_anchor=(0.95, 1.17), frameon=False)


plt.subplots_adjust(hspace = .001)
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[1].spines['right'].set_visible(False)

pois = [8, 29, 45, 57]
for n,poi in enumerate(pois):
    ax[1].axvline(x=poi, ymin=0, ymax=2, c='r', lw=2, clip_on=False)
    ax[0].text(s=n+1, x=poi, y=1.05, c='r')

plt.savefig('competition_verbs.pdf', bbox_inches='tight')
```
