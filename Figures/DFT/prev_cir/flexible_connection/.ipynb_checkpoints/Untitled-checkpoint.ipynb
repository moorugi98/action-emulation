```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.genfromtxt('recording_2022_01_13__10_49_50/r[OUTPUT].boost.csv', skip_header=1, delimiter=',')[:,1]
w = np.genfromtxt('recording_2022_01_13__10_49_50/Hebbian_Connection[BUFFER].weights.csv', skip_header=1, delimiter=',')[:,1]
source = np.genfromtxt('recording_2022_01_13__10_49_50/source[OUTPUT].sigmoided_activation.csv', skip_header=1, delimiter=',')[:,1]
target = np.genfromtxt('recording_2022_01_13__10_49_50/target[OUTPUT].sigmoided_activation.csv', skip_header=1, delimiter=',')[:,1]
```

```python
r.shape
```

```python
plt.rcParams.update({'font.size': 80})
plt.figure(figsize=(20,8))

plt.plot(r, label='reward', linestyle='--', color='black')
plt.plot(w, label='weight', linestyle='-', color='black')
plt.plot(source, label='receiver', linestyle='-', color='blue')
plt.plot(target, label='sender', linestyle='-', color='green')
plt.legend(bbox_to_anchor=(0.95, 1.1), frameon=False)

plt.subplots_adjust(right=2)
plt.xticks([])
plt.xlabel('time')
plt.ylabel('activation')

ymin = 0
ymax = 1.1
plt.vlines(x=13, ymin=ymin, ymax=ymax, color='red')  # all on
plt.vlines(x=23, ymin=ymin, ymax=ymax, color='red')  # reward off
plt.vlines(x=46, ymin=ymin, ymax=ymax, color='red')  # input off, output on
plt.vlines(x=95, ymin=ymin, ymax=ymax, color='red')  # input on, output off
plt.text(s='1', x=13.5, y=1.05, color='red')
plt.text(s='2', x=23.5, y=1.05, color='red')
plt.text(s='3', x=46.5, y=1.05, color='red')
plt.text(s='4', x=95.5, y=1.05, color='red')

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# plt.savefig('time_course_hebb.pdf', bbox_inches='tight')
```
